# **Soy Price Forecast**

This project aims to forecast the price of the brazilian soybeans 60 kilogram bag using only the time series of the price it self. The data was collected from CEPEA (Center for Advanced Studies on Applied Economics) and ranges from 13/03/2006 to 10/04/2025

## Packages and Data


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('data.csv')
```

## Preprocessing

First, i will change the names of the features to english, since the data was aquired from a brazilian source, the names are in portuguese

```python
df = data.rename(columns={'data': 'date', 'valor_real': 'price_BRL', 'valor_dolar' : 'price_USD'})
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst = True)
```

i will set the date as the index since we are working with a time series dataset

```python
missing = df.isnull().sum()
duplicated = df.duplicated().sum()
print(f"Total Missing Values: \n{missing}\nTotal Duplicated Values: {duplicated}")
```

Since we dont have any missing or duplicated values, we can go on to the EDA

## EDA

The price in BRL has a spike in 2020 beacause during the COVID-19 pandemicthe brazilian real has plummeted, with that in mind, is safer for our analysis that we use the price in USD 

```python
df = df.drop(columns=['price_BRL'])
```

### Seasonality Analysis

```python
df_ts = df.set_index('date').sort_index()

result = seasonal_decompose(df_ts['price_USD'], model='multiplicative', period=12)

result.plot()
plt.suptitle('Time Series Decomposition - USD Price', fontsize=16)
plt.tight_layout()
plt.show()
```

# ANALISAR OS RESULTADOS ACIMA

I will move on to use a OLS with monthly dummies to verify if any month has a seazonal impact on the price

```python
df['month'] = df['date'].dt.month
dummies = pd.get_dummies(df['month'], prefix='month', drop_first=True)

X = dummies
y = df['price_USD']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

- Prices tend to be significantly higher in July, August, and September compared to January. This make sense, as the harvest takes place between January and April, scarcity begins at that time
- Durbin-Watson = 0.004 Indicates a strong autocorrelation in residuals. This indicates that it might be a better choice to use SARIMA or Prophet. I will use both and compare the results

```python
df = df.copy()
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
df = df.set_index('date')


sarima_model = SARIMAX(df['price_USD'], order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_result = sarima_model.fit(disp=False)

sarima_pred = sarima_result.get_prediction(start=0, end=len(df)-1)
df['sarima_pred'] = sarima_pred.predicted_mean

df_prophet = df.reset_index().rename(columns={"date": "ds", "price_USD": "y"})
prophet_model = Prophet(yearly_seasonality=True, daily_seasonality=False, weekly_seasonality=False)
prophet_model.fit(df_prophet)

future = prophet_model.make_future_dataframe(periods=0)  # apenas in-sample
forecast = prophet_model.predict(future)

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['price_USD'], label='Observed', color='black', linewidth=1.5)
plt.plot(df.index, df['sarima_pred'], label='SARIMA prediction', color='blue', alpha=0.7)
plt.plot(forecast['ds'], forecast['yhat'], label='Prophet prediction', color='green', alpha=0.7)
plt.title('USD Price - SARIMA vs Prophet (In-sample)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()

prophet_model.plot_components(forecast)
```

```python
sarima_rmse = mean_squared_error(df['price_USD'], df['sarima_pred'], squared=False)
print(f"RMSE (SARIMA): {sarima_rmse}")
prophet_rmse = mean_squared_error(df['price_USD'], forecast['yhat'][:len(df)], squared=False)
print(f"RMSE (Prophet): {prophet_rmse}")
```
